LangGraph Basics

```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    message: str

def greeeting_node(state: AgentState):
    """simple greeting to the user"""
    state['message'] = "Hey " + state['message'] + ", hey how are you?"
    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node('greeter', greeeting_node)
graph.set_entry_point('greeter')
graph.set_finish_point('greeter')
app = graph.compile()
```

```python
from IPython.display import Image, display
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({'message':'bob'})
```

```python
result['message']
```

```python
class NewState(TypedDict):
    name: str

def get_new_name(state: NewState) -> NewState:
    state['name'] = f'hello {state['name']}, you are doing amazing job'
    return state 

graph1 = StateGraph(NewState)
graph1.add_node('welcome', get_new_name)
graph1.set_entry_point('welcome')
graph1.set_finish_point('welcome')
app1 = graph1.compile()

result1 = app1.invoke({'name': 'Bob'})
result1['name']
```
